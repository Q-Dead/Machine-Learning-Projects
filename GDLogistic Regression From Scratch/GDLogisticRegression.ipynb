```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from GDLogReg import LogisticRegression

from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
dataset = datasets.load_breast_cancer()
X, y = dataset.data, dataset.target

X, y = dataset.data, dataset.target 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=46)
```

```python
X_name, y_name = dataset.feature_names, dataset.target_names
```

```python
LogReg = LogisticRegression(learning_rate=0.1, n_iter=10000)
LogReg.fit(np.array(X_train.T[0]).reshape(455,1), y_train)

X_test_reshape = np.array(X_test.T[0]).reshape((114,1)) # Just one feature in order to plot it in 2D plot
log_preg, z = LogReg.predict(np.sort(X_test_reshape.T).T) # for test items just sort it in order to plot it properly
```

```python
np.sort(X_test_reshape.T).tolist()[0][0]
```

```python
category_mapping = {0: y_name[0], 1: y_name[1]}
categorical_labels = [category_mapping[num] for num in y_train]  # Convert to categorical strings
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_train.T[0], categorical_labels, color='#FF1493', label='Data Points', edgecolors='k', alpha=0.5)
plt.plot(np.sort(X_test_reshape.T).T , z, label="Sigmoid function")
plt.xlabel(X_name[0])
plt.ylabel("Diagnosis")
plt.legend()
plt.title("Logistic Regresion of Breast Cancer Diagnosis")
```
