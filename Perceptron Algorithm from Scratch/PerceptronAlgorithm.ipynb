```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from Perceptron import Perceptron_algo
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dt = datasets.load_breast_cancer()
X, y = breast_cancer_dt.data, breast_cancer_dt.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=46)
```

```python
perceptron = Perceptron_algo(learning_rate=0.1, n_iter=10000)
perceptron.fit(X_train, y_train)
y_pred = perceptron.predict(X_test)
```

```python
diagnosis_label = breast_cancer_dt.target_names
diagnosis_value = list(set(y))
diagnosis_label_mapping = { diagnosis_value[0] : diagnosis_label[0], diagnosis_value[1] : diagnosis_label[1],}
```

```python
feature_names = breast_cancer_dt.feature_names
```

```python
y_pred_str_labels = np.array([diagnosis_label_mapping[i] for i in y_pred])
```

```python
x_line = np.linspace(X_test.T[0].min() , X_test.T[0].max() , len(X_test.T[0]))
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_test.T[0], y_pred_str_labels, label="Diagnosis", edgecolors="k", alpha=0.5, s=100)
plt.plot(x_line, np.sort(y_pred_str_labels), c="red")
plt.axhline(0.5, color='black', linestyle='--', label='Threshold')
plt.xlabel(feature_names[0])
plt.legend()
plt.ylabel("Diagnosis")
plt.title("Breast Cancer Diagnosis using Perception Model")
```
