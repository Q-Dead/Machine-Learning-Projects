```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
X_aug = np.ones((len(diabetes_X_train), 1))
```

```python
y_n_dimensional = np.array([diabetes_y_train]).T
X_matrix = np.hstack((X_aug, diabetes_X_train))
```

```python
X_test_aug = np.ones((len(diabetes_X_test), 1))
X_test_matrix = np.hstack((X_test_aug, diabetes_X_test))
```

```python
coef1 = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ y_n_dimensional
y_pred = X_test_matrix @ coef1
```

```python
plt.scatter(diabetes_X_train, diabetes_y_train, marker="x", label="Training Data",alpha=0.4)
plt.scatter(diabetes_X_test, diabetes_y_test, marker="o",c="green", label="Test Data", alpha=1)
plt.plot(diabetes_X_test, y_pred, c="red", label="Prediction")
plt.title('Linear Regression using Normal Equations')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
```
